```python
import os
import shutil
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.yaml")  # load a pretrained model

# Use the model
results = model.train(data='/content/drive/MyDrive/Tumor/config.yaml', epochs=10)  # train the model

```

```python
!scp -r {results.save_dir} '/content/drive/MyDrive/Tumor/runs'
```

```python
from ultralytics import YOLO
# Load a model
model = YOLO("/content/drive/MyDrive/Tumor/runs/weights/best.pt")  # load a custom model

# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.box.map  # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps
```

```python
!scp -r {metrics.save_dir} '/content/drive/MyDrive/Tumor/runs'
```

```python

# Predict with the model
results = model.predict(source="/content/drive/MyDrive/Tumor/pituitary_image.png",save=True, save_txt=True)
```
